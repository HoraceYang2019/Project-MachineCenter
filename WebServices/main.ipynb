## RA_prediction 模型輸入

定位: 商業邏輯轉換 資料儲存

說明: RA_prediction 會select DB中windows窗口的RA相關的特徵，並且將這些特徵組成一個特徵集 (Feature Set)，作為模型的輸入。
並將輸出結果儲存至RA_prediction的table中。

### 輸入特徵集 (Input Features)

| 特徵名稱 | 單位 |說明|
|---------|------|---------|
| `torque_std` | N·m | 刀具扭矩標準差 (Standard Deviation) |
| `torque_mean` | N·m | 刀具扭矩平均值 (Mean) |
| `torque_max` | N·m | 刀具扭矩最大值 (Peak) |
| `spindle_speed_rpm` | RPM | 主軸轉速 (Spindle Speed) |
| `feed_rate_mm_min` | mm/min | 進給速度 (Feed Rate) |


```json
{
  "window_id": "w10233",
  "machine_id": "TMV-720",
  "signals": {
    "torque": {
      "features": {
        "std": 0.079,
        "mean": 2.003,
        "max": 2.08
      }
    },
    "spindle_speed": {
      "features": {
        "mean": 3600
      }
    },
    "feed": {
      "features": {
        "mean": 827.6
      }
    }
  }
}
```

## 模型輸出

### 預測結果 (Prediction Output)

| 欄位 | 類型 | 單位 | 說明 |
|------|------|------|------|
| `prediction_id` | string | - | 預測唯一識別碼 |
| `value` | number | um | 預測的表面粗糙度 Ra 值 |
| `confidence` | number | % | 模型信心度 (0 ~ 1) |
| `model_version` | string | - | 模型版本號（e.g., `rf-ra-v1.2`） |
| `horizon.steps_ahead` | integer | - | 提前預測步數 |
| `horizon.seconds_ahead` | number | s | 提前預測時間 |

```json
{
  "prediction_id": "Prediction_w10233_ra",
  "window_id": "w10233",
  "machine_id": "TMV-720",
  "value": 1.28,
  "unit": "um",
  "confidence": 0.91,
  "model_version": "ra-prediction-v1.2",
  "horizon": {
    "steps_ahead": 1,
    "seconds_ahead": 0.1
  },
  "timestamp": "2026-04-13T10:00:00.400Z"
}
```

## tool_wear_prediction 模型輸入

定位: 商業邏輯轉換 資料儲存

說明: tool_wear_prediction 會select DB中windows窗口的tool_wear相關的特徵，並且將這些特徵組成一個特徵集 (Feature Set)，作為模型的輸入。
並將輸出結果儲存至tool_wear_prediction的table中。

### 輸入特徵集 (Input Features)

| 特徵名稱 | 單位 |說明|
|---------|------|---------|
| `Ap` | mm | 刀具切入深度 (Approach Depth) |
| `Ae` | mm | 刀具切出深度 (Exit Depth) |
| `spindle_speed_rpm` | RPM | 主軸轉速 (Spindle Speed) |
| `feed_rate_mm_min` | mm/min | 進給速度 (Feed Rate) |
| `controller_path` | list | 控制器路徑 (Controller Path) 還需要釐清 |
| `cutting_Temperature` | °C | 切削溫度 (Cutting Temperature) 由模擬計算 |
| `torque` | N·m | 刀具扭矩 (Mean) |
| `bending` | mm | 彎曲量 (Bending) |

```json
{
  "window_id": "w10233",
  "machine_id": "TMV-720",
  "signals": {
    "Ap": {
      "features": {
        "mean": 2.5
      }
    },
    "Ae": {
      "features": {
        "mean": 0.5
      }
    },
    "spindle_speed": {
      "features": {
        "mean": 3600
      }
    },
    "feed": {
      "features": {
        "mean": 827.6
      }
    },
    "controller_path": {
      "features": {
        "path_sequence": [
          "G54 X10 Y20 Z-5",
          "G54 X15 Y25 Z-5",
          "G54 X20 Y30 Z-5"
        ]
      }
    },
    "cutting_Temperature": {
      "features": {
        "mean": 350
      }
    },
    "torque": {
      "features": {
        "mean": 2.003
      }
    },
    "bending": {
      "features": {
        "mean": 0.12
      }
    }
  }
}
```

## 模型輸出

### 預測結果 (Prediction Output)

| 欄位 | 類型 | 單位 | 說明 |
|------|------|------|------|
| `prediction_id` | string | - | 預測唯一識別碼 |
| `tool_wear` | float | % | 預測的刀具磨損程度 |
| `confidence` | number | % | 模型信心度 (0 ~ 1) |
| `model_version` | string | - | 模型版本號 |
| `horizon.steps_ahead` | integer | - | 提前預測步數 |
| `horizon.seconds_ahead` | number | s | 提前預測時間 |

## time_window_aggregator


定位: 彙整 資料儲存

說明:將rowdata中的訊號按照時間窗口進行聚合，計算出每個窗口的統計特徵（如平均值、標準差、最大值等），並將這些特徵存入資料庫中，供後續的模型與分析使用。
透過mqtt_handler接收來自機台的原始數據，time_window_aggregator會根據預設的時間窗口大小（例如每10秒）對數據進行分段，並計算每

## 輸入
```json
{
  "machine_id": "TMV-720",
  "timestamp": "2026-04-13T10:00:00.000Z",
  "signals": {
    "torque": {
      "value": 2.05,
    },
    "bending": {
      "value": 0.15,
    },
    "spindle_speed": {
      "value": 1000,
    },
    "feed": {
      "value": 500,
    },
    "STH_power": {
      "value": 8.11,
    },
    "STH_RSSI": {
      "value": -70,
    },
    "STH_TTOLP": {
      "value": 60,
    },
    "position": {
      "x": 10.5,
      "y": 20.3,
      "z": -5.0
    }
  }
}
```

## 輸出

```json
{
  "window_id": "w10233",
  "machine_id": "TMV-720",
  "signals": {
    "torque": {
      "features": {
        "std": 0.079,
        "mean": 2.003,
        "max": 2.08
      }
    },
    "bending": {
      "features": {
        "std": 0.02,
        "mean": 0.12,
        "max": 0.15
      }
    },
    "spindle_speed": {
      "features": {
        "mean": 3600
      }
    },
    "feed": {
      "features": {
        "mean": 827.6
      }
    },
    "STH_power": {
      "features": {
        "mean": 8.11
      }
    },
    "STH_RSSI": {
      "features": {
        "mean": -70
      }
    },
    "STH_TTOLP": {
      "features": {
        "mean": 60
      }
    },
    "position": {
      "features": {
        "x_mean": 10.5,
        "y_mean": 20.3,
        "z_mean": -5.0
      }
    }
  }
}
```

## mqtt_handler

定位:real-time data processing ui顯示

說明: mqtt_handler 負責接收來自機台的 MQTT 訊息，將數據攔截並輸入至瀏覽器系統中，給予torque bending顯示在瀏覽器上。
啟動服務就會監聽MQTT broker的特定topic，當有新的訊息發布時，mqtt_handler會將數據輸入至瀏覽器系統中，供前端UI使用。

## 無須輸入

## 輸出

```json
{
  "machine_id": "TMV-720",
  "timestamp": "2026-04-13T10:00:00.000Z",
  "signals": {
    "torque": {
      "value": 2.05,
    },
    "bending": {
      "value": 0.15,
    },
    "spindle_speed": {
      "value": 1000,
    },
    "feed": {
      "value": 500,
    },
    "STH_power": {
      "value": 8.11,
    },
    "STH_RSSI": {
      "value": -70,
    },
    "STH_TTOLP": {
      "value": 60,
    },
    "position": {
      "x": 10.5,
      "y": 20.3,
      "z": -5.0
    }
  }
}
```


## get_window_data 

定位: 資料查詢 ui顯示

說明: 將歷史資料降取樣後顯示在UI上，供使用者查看特定時間範圍內的機台數據趨勢。

輸入
```json
{
  "machine_id": "TMV-720",
  "tool_id": "tool_123",
  "start_window_id": "w10231",
  "end_window_id": "w10232",
  "signals": ["torque", "bending"]
}
```

輸出 

```json
{
  "machine_id": "TMV-720",
  "tool_id": "tool_123",
  "window_id": ["w10231", "w10232"],
  "signals": {
    "torque": {
      "values": [2.0, 2.1],
    },
    "bending": {
      "values": [0.1, 0.12],
    }
  }
}
```
